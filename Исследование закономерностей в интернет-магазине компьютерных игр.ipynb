# Исследование закономерностей в интернет-магазине компьютерных игр.

Интернет-магазин «Стримчик» осущствляет продажи компьютерных игр по всему миру. 

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 


В нашем распоряжении данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Цель исследования:**

Выявление закономерностей, определяющих успешность игры, что позволит сделать ставку на потенциально популярный продукт и спланировать прибыльную рекламную компанию.

**Описание данных:**

 - `Name` — название игры
    
 - `Platform` — платформа

 - `Year_of_Release` — год выпуска

 - `Genre` — жанр игры 

 - `NA_sales` — продажи в Северной Америке (миллионы проданных копий)

 - `EU_sales` — продажи в Европе (миллионы проданных копий)

 - `JP_sales` — продажи в Японии (миллионы проданных копий)

 - `Other_sales` — продажи в других странах (миллионы проданных копий)

 - `Critic_Score` — оценка критиков (максимум 100)

 - `User_Score` — оценка пользователей (максимум 10)

 - `Rating` — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.


Данные за 2016 год могут быть неполными.

```python
def platforms(name_of_platform):

    platform = actual_date.loc[actual_date['platform'] == name_of_platform]
    fig, ax = plt.subplots(1, 2, figsize=(15,5))
    sns.scatterplot(x='user_score', y='sum_sales', data=platform, ax=ax [0])
    sns.scatterplot(x='critic_score', y='sum_sales', data=platform, ax=ax [1])
    fig.suptitle(name_of_platform, fontsize=15)
    ax[0].set(xlabel = 'Оценка пользователей')
    ax[1].set(xlabel = 'Оценка критиков')
    ax[0].set(ylabel = 'Продажи, млн. копий')
    ax[1].set(ylabel = 'Продажи, млн. копий')
    plt.show()
    
    correl = platform['user_score'].corr(platform['sum_sales'])
    critic_correl = platform['critic_score'].corr(platform['sum_sales'])
    
    print('Оценка пользователей', correl)
    print('Оценка критиков', critic_correl)
```

## Проверка гипотез

### Гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Сформулируем нулевую и альтернативную гипотезы

 - **H0** - средние пользовательские рейтинги по платформам Xbox One и PC одинаковые

 - **H1** - средние пользовательские  рейтинги по платформам Xbox One и PC разные.

```python
#фильтруем данные и удаляем пропуски
xone = actual_date.loc[actual_date['platform'] == 'XOne']['user_score'].dropna()
pc = actual_date.loc[actual_date['platform'] == 'PC']['user_score'].dropna()
```

```python
#Считаем средний пользовательский рейтинг для платформы XOne
xone.mean()
```

```python
#Считаем средний пользовательский рейтинг для платформы PC
pc.mean()
```

Визуализируем гистограммы пользовательских рейтингов для платформ, а также средний пользовательский рейтинг.

```python
#Задаем сетку отрисовки графиков
plt.figure(figsize=(14, 5))


ax1 = plt.subplot(1, 2, 1)
sns.histplot(xone, bins=30, ax=ax1) 
plt.axvline(xone.mean(), color='red', linestyle='dashed', linewidth=2)
plt.xlabel('Пользовательский рейтинг')
plt.ylabel('Количество игр')
plt.title('Гистограмма пользовательского рейтинга для XOne')


ax2 = plt.subplot(1, 2, 2)
sns.histplot(pc, bins=30, ax=ax2) 
plt.axvline(pc.mean(), color='red', linestyle='dashed', linewidth=2)
plt.xlabel('Пользовательский рейтинг')
plt.ylabel('Количество игр')
plt.title('Гистограмма пользовательского рейтинга для PC')

plt.show()
```

Посчитаем размер выборок

```python
print('Размер выборки пользовательского рейтинга платформы XOne:', xone.count())
print('Размер выборки пользовательского рейтинга платформы PC:', pc.count())
```

Для проверки гипотезы используем двусторонний st.ttest_ind() так как выборки можно считать независмыми, меющие более 30 значений, поэтому нормальность выборки не важна, с учетом того, что в реальных бизнес-процессах нормального распределения данных ни когда не бывает и близкими к нормальному распределению, поэтому применен двусторонний t-test. Выборки разные по размеру, дисперсии, предположительно, неоднородные, поэтому параметр equal_var=False

```python
# задаём уровень значимости
alpha = 0.05

# проводим двусторонний ttest для двух выборок
results = st.ttest_ind(xone, pc, equal_var=False)

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

**Вывод:** 

На имеющихся данных, на уровне значимости 5% нет оснований отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) платформ  Xbox One и PC  в пользу альтернативной гипотезы.

### Гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

Сформулируем нулевую и альтернативную гипотезы. Нулевая гипотеза всегда формулируется на равенство, поэтому за нулевую гипотезу возьмем, что средние пользовательские рейтинги жанров Action и Sports равны, а альтернативная гипотеза будет звучать, что средние пользовательские рейтинги жанров разные.

 - **H0** - средние пользовательские рейтинги жанров Action и Sports одинаковые.

 - **H1** - средние пользовательские рейтинги жанров Action и Sports разные.

```python
#фильтруем данные и удаляем пропуски
action = actual_date.loc[actual_date['genre'] == 'Action']['user_score'].dropna()
sports = actual_date.loc[actual_date['genre'] == 'Sports']['user_score'].dropna()
```

```python
#вычислим средние значения по выборкам
```

```python
action.mean()
```

```python
sports.mean()
```

Визуализируем гистограммы пользовательских рейтингов для жанров, а также средний пользовательский рейтинг по жанрам.

```python
#Задаем сетку отрисовки графиков
plt.figure(figsize=(14, 5))

ax1 = plt.subplot(1, 2, 1)
sns.histplot(action, bins=30, ax=ax1) 
plt.axvline(action.mean(), color='red', linestyle='dashed', linewidth=2)
plt.xlabel('Пользовательский рейтинг')
plt.ylabel('Количество игр')
plt.title('Гистограмма пользовательского рейтинга для жанра Action')

ax2 = plt.subplot(1, 2, 2)
sns.histplot(sports, bins=30, ax=ax2) 
plt.axvline(sports.mean(), color='red', linestyle='dashed', linewidth=2)
plt.xlabel('Пользовательский рейтинг')
plt.ylabel('Количество игр')
plt.title('Гистограмма пользовательского рейтинга для жанра Sports')

plt.show()
```

На визуализации пользовательские рейтинги для жанров Action и Sports различны.

```python
print('Размер выборки пользовательского рейтинга жанра Action:', action.count())
print('Размер выборки пользовательского рейтинга жанра Sports', sports.count())
```

Для проверки гипотезы используем двусторонний st.ttest_ind() так как выборки можно считать независмыми, имеющие более 30 значений, поэтому нормальность выборки не важна, с учетом того, что в реальных бизнес-процессах нормального распределения данных ни когда не бывает и близкими к нормальному распределению, поэтому применен двусторонний t-test. Выборки разные по размеру, дисперсии, предположительно, неоднородные, поэтому параметр `equal_var=False`

```python
# задаём уровень значимости
alpha = 0.05

# проводим двусторонний ttest для двух выборок
results = st.ttest_ind(action, sports, equal_var=False)

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```

**Вывод:** 

На имеющихся данных, на уровне значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов (отзывов) жанров Action и Sports в пользу альтернативной гипотезы.
